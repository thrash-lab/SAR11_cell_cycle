```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import pandas as pd
import numpy as np

def create_ipr_dataframe(text):
  """
  Transforms a string of InterPro accessions and descriptions into a pandas DataFrame.

  Args:
    text: A string where each line contains an InterPro accession followed by a space and its description.

  Returns:
    A pandas DataFrame with two columns: 'IPR_accession' and 'Description'.
  """
  # Split the text into lines and then split each line by the first space
  data = [line.split(' ', 1) for line in text.strip().split('\n')]

  # Create a pandas DataFrame
  output_ipr_df = pd.DataFrame(data, columns=['IPR_accession', 'Description'])

  return output_ipr_df


def plot_interpro_heatmap(text, gene_cluster_df, iprscan_df, phylogenomic_order, title):
    """
    Generates a heatmap of InterPro hits per genome.

    Args:
        text: A string where each line contains an InterPro accession followed by a space and its description.
        gene_cluster_df: DataFrame containing gene cluster information with 'genome_name' and 'aa_sequence'.
        iprscan_df: DataFrame containing InterPro scan results with 'Protein_accession_index' and 'InterPro annotations - accession'.
        phylogenomic_order: DataFrame with the desired order of genomes in the heatmap.

    Returns:
        A pandas DataFrame representing the hit matrix.
    """
    # Create the DataFrame of InterPro accessions and descriptions
    def create_ipr_dataframe(text):
        data = [line.split(' ', 1) for line in text.strip().split('\n')]
        return pd.DataFrame(data, columns=['IPR_accession', 'Description'])

    output_ipr_df = create_ipr_dataframe(text)

    # Create the hit matrix
    hit_matrix = []
    for IPR_accession in output_ipr_df['IPR_accession'].values:
        hit_array = np.zeros(len(phylogenomic_order['Genome name']))
        # Ensure 'Protein_accession_index' is treated as integers for unique()
        accession_indices = iprscan_df[iprscan_df['InterPro annotations - accession'] == IPR_accession]['Protein_accession_index'].astype(int).unique()
        hit_genomes = gene_cluster_df.loc[accession_indices, 'genome_name'].values
        for hg in hit_genomes:
             # Ensure the genome name exists in phylogenomic_order before trying to get its index
            if hg in phylogenomic_order['Genome name'].values:
                hit_array[phylogenomic_order[phylogenomic_order['Genome name'] == hg].index] = 1


        hit_matrix.append(hit_array)

    gene_hit_df = pd.DataFrame(hit_matrix, columns=phylogenomic_order['Genome name'].values, index=output_ipr_df['IPR_accession'].values).T

    # Create the heatmap
    plt.figure(figsize=(10, 15)) # Reverted figure size
    ax = sns.heatmap(gene_hit_df, cmap="YlGnBu", cbar=True, yticklabels=False) # Removed ax argument

    plt.title(title)
    plt.xlabel('InterPro Accession')
    plt.ylabel('Genome Name')
    # Removed plt.tight_layout()
    plt.show()

    return gene_hit_df

# Example usage of the function:
text = """IPR000158 Cell_div_FtsZ
IPR020805 Cell_div_FtsZ_CS
IPR045061 FtsZ/CetZ
IPR024757 FtsZ_C
IPR008280 Tub_FtsZ_C
IPR037103 Tubulin/FtsZ-like_C
IPR018316 Tubulin/FtsZ_2-layer-sand-dom
IPR036525 Tubulin/FtsZ_GTPase_sf
IPR003008 Tubulin_FtsZ_GTPase
"""

# Assuming gene_cluster_df, iprscan_df, and phylogenomic_order are already loaded
# You would call the function like this:
gene_hit_dataframe = plot_interpro_heatmap(text, gene_cluster_df, iprscan_df, phylogenomic_order, 'FtsZ')
```
